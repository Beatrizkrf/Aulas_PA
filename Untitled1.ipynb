Escreva um programa que solicite um número inteiro e positivo pra o usuario e na sequência o programa deve calcular um valor como mostrado no exemplo abaixo:

numero digitado: 4

valor = 1/1 + 1/2 + 1/ 3+ 1/4

valor= 2,08

```python
numero = (int(input('Digite um número POSITIVO e INTEIRO desejado: ')))
valor= 0
for i in range(1,numero+1):
  valor = valor + 1/i
print('Valor total: ',valor)
```
