## Faça um programa que receba um númer inteiro digitado pelo usuario e diga se o número é primo ou não:

```python
numero= int(input("Digite o número desejado: "))
quant=0
if (numero < 0):
  print("Número inválido, deve ser positivo")
else:
  if(numero == 0) or (numero==1):
    print("Por definição 0 ou 1 não é primo!")
  else:
    for i in range(1,numero+1):
   #print(i)
      resto = numero % i
      print(f'resto da divisão de {numero} por {i} = {resto}')
      if resto == 0:
        quant +=1
    print('Qunatidade de restos igaul a zero',quant)
    if (quant ==2):
      print(f'O número {numero} é primo')
    else:
      print(f'O número {numero} não é primo')
```
