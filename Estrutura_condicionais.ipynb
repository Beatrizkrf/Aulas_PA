# **Estrura Condicionais**

## Simples:

```python
#avaliar se um aluno foi aprovado caso sua média de 3 notas for maior ou igual a 5
#Entrada das 3 notas
n1= float(input('Digite a primeira nota:'))
n2= float(input('Digite a segunda nota:'))
n3= float(input('Digite a terceira nota:'))
```

```python
#processamento da média
media= (n1+n2+n3)/3
media
```

```python
#avaliando o valor da média se é maior ou igual a 5
if(media>=5):
  print('Aprovado')
print('Fim do programa...')
```

## Composto:

```python
#avaliar se um aluno foi aprovado ou reprovado caso sua média de 3 notas for maior ou igual a 5
#Entrada das 3 notas
n1= float(input('Digite a primeira nota:'))
n2= float(input('Digite a segunda nota:'))
n3= float(input('Digite a terceira nota:'))

#processamento da média
media= (n1+n2+n3)/3
media
```

```python
if(media>=5):
  print('Aprovado')
else:
  print('Reprovado')
print('Fim do programa...')
```

## Encadeado:

```python
#avaliar se um aluno foi aprovado ou reprovado caso sua média de 3 notas for maior ou igual a 5
#aprovado media > 5
#reprovado medias < 4
#recuperação entre 4 e 5

#Entrada das 3 notas
n1= float(input('Digite a primeira nota:'))
n2= float(input('Digite a segunda nota:'))
n3= float(input('Digite a terceira nota:'))

#processamento da média
media= (n1+n2+n3)/3
media
```

```python
if media>5:
  print('Aprovado')
else:
  if media<4:
    print('Reprovado')
  else:
      print('Recuperação')
print('Fim do programa...')
```

```python
#outra forma de resolução é colocando o 'elif' em vez de colocar novamente o if e o else
if media>5:
  print('Aprovado')
elif media<4:
    print('Reprovado')
else:
      print('Recuperação')
print('Fim do programa...')
```
