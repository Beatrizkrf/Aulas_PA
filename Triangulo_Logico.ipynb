# **Triângulo Lógico:**

```python
a=float(input('Digite o lado A do triângulo:'))
b=float(input('Digite o lado B do triângulo:'))
c=float(input('Digite o lado C do triângulo:'))

triangulo=((a+b>c)and (b+c>a) and (a+c>b))
print('É um triangulo?',triangulo)
```

Equilatero


```python
equilatero=((a==b) and (b==c)and (triangulo))
print('O triangulo é equilatero?',equilatero)
```

Escaleno

```python
escaleno= ((a!=b) and (b!=c) and (c!=a)and (triangulo))
print('O triangulo é escaleno?',escaleno)
```

Isosceles

```python
isosceles= ((a!=b) or (b!=c) or (c!=a) or (triangulo))
print('O triangulo é isosceles?',isosceles)
```
