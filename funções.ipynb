# **Função sem retorno e com parametros**





```python
def potencia(base, expoente):
  resultado= base**expoente
  print("o resultado é: "+ str(resultado))
potencia(2,3)
```

```python
def raiz (num):
  raiz= num**(0.5)
  print("a raiz quadrada é igual a: ",raiz)
raiz(9)
```

```python
def somar(x,y):
  soma= x+ y
  return soma

print("exemplo com retorno")
num1 = int(input("digite o primeiro numero: "))
num2 = int(input("digite o segundo numero: "))
retorno = somar(num1, num2)
print("voltou da função ", retorno)
```

Com base no exemplo anterior, cite um programa que peça três vezes dois números e utilize a função acima para soma-los e exibir na tela o resultado.

```python
def somar(x,y):
  soma= x+ y
  return soma

for i in range(3):
  num1 = int(input("digite o primeiro numero: "))
  num2 = int(input("digite o segundo numero: "))
  retorno = somar(num1, num2)
  print("voltou da função ", retorno)
```
